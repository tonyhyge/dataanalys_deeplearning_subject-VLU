13. Thay thế các giá trị của cột Age> lớn hơn 40 bằng giá trị trung bình của cột đó

14. Biết công thức tính chỉ số BMI như sau:

```python
data["cs_BMI"]=data.apply(lambda x: (x.Weight/(x.Height**2)), axis=1)
data
```

15. Theo thang đo kết quả tính BMI của châu Á:

```python
data['Phan_Loai'] = pd.cut(data['cs_BMI'], bins=[-float('inf'), 18.5, 23, 25, float('inf')],
                         labels=['0', '1', '2', '3'])
```

```python
data
```

16. Chuẩn hóa giá trị của cột cs_BMI> chuẩn hóa giá trị trong đoạn từ 0 tới 1.

```python
data['cs_BMI'] = (data['cs_BMI'] - data['cs_BMI'].min()) / (data['cs_BMI'].max() - data['cs_BMI'].min())
data
```

17. Cho biết:
+ ‘Mr‘tương ứng với ‘male’,
+ ’Mrs’ và ‘Miss’ tương ứng với ‘female’.
Tạo cột với tên GT>, Sau khi thực hiện, xóa các từ Mr, Mrs, Miss nếu trong
cột Name> còn các từ đó.

```python
data['GT'] = data['Name'].apply(lambda x: 'male' if 'Mr' in x else ('female' if 'Mrs' in x or 'Miss' in x else x))
data['GT'] = data['GT'].apply(lambda x: x.replace('Mr', '').replace('Mrs', '').replace('Miss', ''))

data['Name'] = data['Name'].str.replace('Mr.', '')
data['Name'] = data['Name'].str.replace('Mrs.', '')
data['Name'] = data['Name'].str.replace('Miss.', '')

```

```python
data
```
